## Check the number of rentals for each customer for May.


```python
query7 = '''
        SELECT m.customer_id, first_name, last_name, count(rental_id) as Number_Rents_may, rank() over(order by count(rental_id) desc) as 'rank'
        FROM customer c
        JOIN may_rentals m
        ON m.customer_id = c.customer_id
        GROUP BY m.customer_id
    
        '''

rentals_per_customer_may = pd.read_sql_query(query7, engine)
rentals_per_customer_may
```

## Check the number of rentals for each customer for June.
 

```python
query8 = '''
        SELECT j.customer_id, first_name, last_name, count(rental_id) as Number_Rents_june, rank() over(order by count(rental_id) desc) as 'rank'
        FROM june_rentals j
        JOIN customer c
        ON j.customer_id = c.customer_id
        GROUP BY j.customer_id
    
        '''

rentals_per_customer_june = pd.read_sql_query(query8, engine)
rentals_per_customer_june
```

```python
#replace NAN with 0 in columns of rental counts may and june
rental_count_0506['Number_Rents_june'] = rental_count_0506['Number_Rents_june'].fillna(0)
rental_count_0506['Number_Rents_may'] = rental_count_0506['Number_Rents_may'].fillna(0)
rental_count_0506
```

```python
def compare_rentals(i):
        if i['Number_Rents_june']<i['Number_Rents_may']:
            print ("Customer", i['customer_id'], "has more rentals in May.")
        elif i['Number_Rents_june'] > i['Number_Rents_may']:
            print ("Customer", i['customer_id'], "has more rentals in June.")
        elif i['Number_Rents_june'] == i['Number_Rents_may']:
            print ("Customer", i['customer_id'], "has more the same amount of rentals in May and June.")
        else:
            print ("Customer", i['customer_id'], "'s rentals cannot be compared.")
```

```python
#apply function for each row (axis=1)
rental_count_0506.apply(compare_rentals, axis=1)
```
